```python
import pandas as pd
import numpy as np
# import umap
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import ast
```

```python
# CSV 파일 불러오기
file_path = "/content/human_feedback_embedding_small.csv"
df = pd.read_csv(file_path)
```

```python
# 문자열을 리스트로 변환후, numpy 배열로 변환
df["embedding"] = df["embedding"].apply(ast.literal_eval)
embeddings = np.array(df["embedding"].tolist())
```

```python
# umap 통해서 차원 축소
# umap_reducer = umap.UMAP(
#     n_neighbors =3,
#     min_dist = 0.1,
#     n_components = 50,
#     random_state = 42,
#     metric = "cosine"
# )

# umap_embeddings_50d = umap_reducer.fit_transform(embeddings)
```

```python
#DBSCA 적용
dbscan = DBSCAN(
    eps = 0.2,
    min_samples = 5,
    metric = "cosine"
)
clusters = dbscan.fit_predict(embeddings)
```

```python
df["cluster"]=clusters
```

```python
#클러스터별 데이터 개수
cluster_counts = df["cluster"].value_counts()
print("클러스터별 개수:/n", cluster_counts)
```
